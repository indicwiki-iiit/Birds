```python
import pandas as pd
eight_data_sets = [
    pd.read_pickle('Pick_ebird_1_1500.pkl'),
    pd.read_pickle('Pick_ebird_1500_3000.pkl'),
    pd.read_pickle('Pick_ebird_3000_4500.pkl'),
    pd.read_pickle('Pick_ebird_4500_6000.pkl'),
    pd.read_pickle('Pick_ebird_6000_7500.pkl'),
    pd.read_pickle('Pick_ebird_7500_9000.pkl'),
    pd.read_pickle('Pick_ebird_9000_10500.pkl'),
    pd.read_pickle('Pick_ebird_10500_11187.pkl'),]

ebird_dataset = pd.concat(eight_data_sets,sort=False)
ebird_dataset.shape
```

```python
ebird_dataset.to_csv('ebird_data.csv')
ebird_dataset.to_pickle('ebird_data.pkl')
ebird_dataset.to_excel('Merged_Ebird.xlsx')
```
