```python
import pandas as pd
import numpy as np
```

```python
iucn_data_sets = [
    pd.read_pickle('Pick_iucn_1_2790.pkl'),
    pd.read_pickle('Pick_iucn_2790_5580.pkl'),
    pd.read_pickle('Pick_iucn_5580_8370.pkl'),
    pd.read_pickle('Pick_iucn_8370_11163.pkl')
    ]

iucn_dataset = pd.concat(iucn_data_sets,sort=False)
iucn_dataset.shape
```

```python
# iucn_dataset.index
iucn_dataset.head(5)
```

```python
import re

def clean_sci_year(text):
    if type(text) == str:
        text = text.strip()
        text = re.sub('\n', '', text)
        text = re.sub('\(', '', text)
        text = re.sub('\)', '', text)
        text = text.strip()
        text = re.sub('[-‘’“”…]', '', text)
        text = text.split(',')
        text = [_.strip() for _ in text if _!='']
        if len(text) != 2:
            if len(text) > 2:
                return [','.join(text[:-1]),text[-1]]
            return [np.nan,np.nan]
        else:
            return text
    return [np.nan,np.nan]

# clean = lambda x: clean_text(x)
```

```python
k = [i for i in range(iucn_dataset.shape[0])]
iucn_dataset['index'] = k
iucn_dataset.set_index('index',inplace=True)
iucn_dataset.head(4)
iucn_dataset.shape[0]
```

```python
from tqdm import tqdm
# iucn_dataset['Scientist'] = np.nan
# iucn_dataset['founded'] = np.nan
d = pd.DataFrame(columns=['Name','Scientist','founded'])
scientist_names = []
found_years = []
for i in tqdm(range(iucn_dataset.shape[0])):
# for i in tqdm(range(5)):
    scientist_name,found_year = clean_sci_year(iucn_dataset.loc[i,'sci_n_year'])
    name = iucn_dataset.loc[i,'Name']
    d.loc[i] = [name,scientist_name,found_year]
    # iucn_dataset[i,'Scientist'] = scientist_name
    # iucn_dataset[i,'founded'] = found_year
    # scientist_names.append(scientist_name)
    # found_years.append(found_year)
# print(scientist_names)
# iucn_dataset =  pd.concat([iucn_dataset,pd.DataFrame({'Scientist':scientist_names,'Founded Year':found_years})])
# iucn_dataset['Scientist']=scientist_names
# iucn_dataset['Finded In'] = found_years
iucn_dataset.head(5)
```

```python
iucn_dataset = pd.merge(iucn_dataset,d)
```
