```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from anuvaad import Anuvaad
anu = Anuvaad("english-telugu")
# se = "/kaggle/input/birds-data/data_v2_1_geographic_distribution_includes_anuvaad.xlsx"
# data = pd.read_pickle("/content/data_v5_habitat_anuvaad.pkl")
def create_telugu_col_new_optim(col,data):
    di = {}
    # data[col+ ' Telugu'] = np.nan
    for i in tqdm(range(data.shape[0])):
        # print(i)
        if type(data[col][i]) == float:
            data[col + ' Telugu'][i] = np.nan
        else:
            if data[col][i] in di:
                data[col + ' Telugu'][i] = di[data[col][i]]
            else:
                data[col + ' Telugu'][i] = anu.anuvaad(data[col][i])
                di[data[col][i]] = data[col + ' Telugu'][i]
    return data
```

```python
data = create_telugu_col_new_optim('Bird Original Name',data)
out(data,'3_1')
```

```python
data = create_telugu_col_new_optim('Scientist',data)
out(data,'3_2')
```
