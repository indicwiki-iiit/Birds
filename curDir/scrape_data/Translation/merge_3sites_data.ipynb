```python
import re
def split_upper(s):
    return list(filter(None, re.split("([A-Z][^A-Z]*)", s)))
```

```python
fail = []
def clean_bn(cn,wiki):
    try:
        # print('h')
        bn = wiki.loc[cn,'Binomial Name']
        # print(bn)
        if '†' in bn:
            wiki.loc[cn,'Binomial Name'] = bn.replace('†','')
        if '(' in bn:
            # print(wiki.loc[cn,'Binomial Name'].split(' ')[0] + wiki.loc[cn,'Binomial Name'].split(' ')[1].split('(')[0])
            wiki.loc[cn,'Binomial Name'] = wiki.loc[cn,'Binomial Name'].split(' ')[0] +' '+ wiki.loc[cn,'Binomial Name'].split(' ')[1].split('(')[0]
        else:
            # print(wiki.loc[cn,'Binomial Name'])
            wiki.loc[cn,'Binomial Name'] = wiki.loc[cn,'Binomial Name'].split(' ')[0] +' '+ split_upper(wiki.loc[cn,'Binomial Name'].split(' ')[1])[0]
    except Exception:
        # print('Failed :',cn)
        fail.append(cn)
    finally:
        return wiki
        # pass

# clean_bn('Abyssinian crimsonwing',wiki).head(5)
# clean_bn('Abyssinian longclaw',wiki)

```

```python
for i in list_of_bird_names:
    wiki = clean_bn(i,wiki)
wiki.head(10)
```

```python
def count_col(wiki):
    wiki_total = wiki.shape[0]
    wiki_d = {}
    for i in wiki.columns:
        wiki_d[i] = wiki_total- wiki[i].isnull().sum()
    wiki_d = sorted(wiki_d.items(), key=lambda x:x[1],reverse=True)
    wiki_df = pd.DataFrame(wiki_d,columns=['Attribute','Occur'])
    wiki_df['Perc'] = (wiki_df['Occur']/wiki_total)*100  
    return wiki_df
count_col(wiki).to_csv('../assets/wiki_col_count.csv')
```
