### Clean : Wikipedia.org data

```python
wiki_merged = pd.read_csv('../assets/Wiki_Merged.csv',index_col='Common Name')
wiki_merged.drop(['ID'],axis=1,inplace=True)
```

```python
import pickle
import re
import string

def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\n', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('[-‘’“”…]', '', text)
    return text

clean = lambda x: clean_text(x)
```

```python
def new_clean_text(text):
    ''' remove Spaces '''
    text = text.lower()
    text = re.sub(' ', '', text)
    return text

clean_space = lambda x: new_clean_text(x)
```

```python
import numpy as np
def merge_maha(df,A,B,C,defa=False):
    if defa:
        defa = B
    else:
        defa = A
    m1 = ~df[A].isna()
    # df['m1'] = m1
    m2 = ~df[B].isna()
    # df['m2'] = m2
    df[C] = np.select([m1, m2, m1 | m2],[df[A],df[B],df[A]],default=df[defa])
    # df = df.drop([A,B])
    return df
```

```python
# names_ebird_wiki = merge_maha(names_ebird_wiki,'Kingdom_x','Kingdom_y','Kingdom')
# merge_maha(names_ebird_wiki,'Kingdom_x','Kingdom_y','Kingdom')
names_ebird_wiki = merge_maha(names_ebird_wiki,'Family_x','Family_y','Family')
names_ebird_wiki = merge_maha(names_ebird_wiki,'Kingdom_x','Kingdom_y','Kingdom')
names_ebird_wiki = merge_maha(names_ebird_wiki,'Phylum_x','Phylum_y','Phylum')
names_ebird_wiki = merge_maha(names_ebird_wiki,'Class_x','Class_y','Class')
names_ebird_wiki = merge_maha(names_ebird_wiki,'Order_x','Order_y','Order')
```
